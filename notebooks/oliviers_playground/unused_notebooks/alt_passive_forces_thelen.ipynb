```python
import numpy as np
import tensorflow as tf
import matplotlib.pyplot as plt


fttoe = 0.33
eps0 = 0.04
# to make the slope of the linear section equal to 42.8 like in Thelen et al. 2003
x = (2 * eps0 * fttoe) / (42.8 * eps0 ** 2)
eps_toe = eps0 * x
k_se = fttoe / eps_toe ** 2
a = k_se * 2 * eps_toe


eps = np.linspace(0., 0.06, 101)
f = tf.where(eps > eps_toe, a * eps - fttoe, k_se * eps ** 2)
plt.plot(eps, f)
plt.show()

```
